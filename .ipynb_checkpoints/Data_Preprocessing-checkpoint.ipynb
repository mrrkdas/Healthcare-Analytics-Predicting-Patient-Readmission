```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("hospital_readmissions.csv")
df.head()
```

```python
df.isna()
```

```python
unique_values = df["medical_specialty"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_vars = np.array(df[["medical_specialty", "diag_1", "diag_2", "diag_3"]])
```

```python
encoder = LabelEncoder()

new_categorical_values = encoder.fit_transform(categorical_vars.reshape((-1,1)))
```

```python
df["medical_specialty", "diag_1", "diag_2", "diag_3"] = new_categorical_values
```

```python
cat_columns = df.select_dtypes(['category']).columns
```

```python
cat_columns
```

```python
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```
